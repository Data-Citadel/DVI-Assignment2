### Question 01 [3 marks]

**A sample from the salaries data is provided in the file "ex01.txt".** 

Plot and appropriate graph to help you derive an inference. You will have to explicitly state your inference what you have found.

Loading the data from `ex01.txt` file.

```python
cols = ["Salary","Experience"]

data_df = pd.read_csv("ex01.txt", names=cols, header=None);

data_df.info()
print("=="*40)
data_df.head()
```

```python
plt.figure(figsize=(8,7))

plt.scatter(data_df.Salary, data_df.Experience)

plt.xlabel("Salary Range")
plt.ylabel("Years of Experience")
plt.title("Scatter plot", fontsize=14)

plt.show()
```

```python
f = plt.figure(figsize=(8, 7))
ax = f.gca()
sns.regplot(x="Salary", y="Experience", data=data_df, ax=ax); # Regression line plot
ax.set_title("Regression line between X and Y", fontsize=14)
plt.show()
```

### Inference: 
    From the given data and the above visualization, it is clear that the salary and the number of years of experience have a strong relation, as the years of experience increases, the salary also increases. We infer that the data attributes are positively related.

### Question 2 [3 marks]

**Analyze the GDP of the country from the data in the file "ex02.txt".**

Use the appropriate plot to help you derive an inference. You will have to explicitly state your inference what you have found.

Loading the data from `ex02.txt` file.

```python
cols = ['year', 'gdp']

ex2_df = pd.read_csv("ex02.txt", names=cols, header=None);

ex2_df.info()
print("=="*30)
ex2_df.head()
```

```python
fig = plt.figure(figsize=(8,7))
ax = fig.gca()

sns.lineplot(ex2_df['year'], ex2_df['gdp'], linewidth=2.5, ax=ax)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('GDP growth', fontsize=12)
ax.set_title("Showing the Trend using Line chart", fontsize=16)

plt.show()
```

### Inference: 
    
    The above plot depicts the GDP growth data against years from 2011 to 2019. The plot explains the following key points,
    1. There was a decline in the GDP from 2011 to 2012.
    2. The trend was changed and there was a steady increase in the GDP until 2016
    3. Followed by a stationary momentum till 2017. 
    4. From 2017 to 2018 there was a decline in the line and again there was increase in the GDP. 
    

### Question 3 [3 marks]

**The data of Oxygen and Carbon Levels of 2 cities over a period of time are given in the file "ex03.txt".** 

Use the appropriate plot to help you derive an inference. You will have to explicitly state your inference what you have found.

```python
ex3_df = pd.read_csv("ex03.txt", header=0);

ex3_df.info()
print("=="*30)
ex3_df.head()
```

```python
ax1 = sns.catplot(x="Year", y="O2", hue="City", data=ex3_df, kind="bar", palette='bright')
ax1.set_ylabels("Oxygen level in cities")
ax2 = sns.catplot(x="Year", y="CO2", hue="City", data=ex3_df, kind="bar", palette="bright")
ax2.set_ylabels("Carbon di-oxide level in cities")
plt.show()
```

### Inference:   

The given dataset had information on the cities O<sub>2</sub> and CO<sub>2</sub> across different years. We used a stacked bar chart to find the levels of O<sub>2</sub> and CO<sub>2</sub>. 
We can observe that the city *Delhi* has higher volume in both Oxygen and Carbon di-oxide. With respect to the Oxygen, both the cities are more or less similer, whereas with respect to CO2 Bangalore is doing great.

#### The number of tweets done for hashtag #dvicourse is provided in the file "ex05.txt" file. Use the appropriate  plot to display them in form of 4 quartiles 0-25, 26-50, 51-75, 75-100and derive an inference. State explicitly the inference you have derived.

```python
df = pd.read_csv("ex05.txt", header=None);
df.columns = ["Tweets"]
df.head()
```

```python
df.Tweets.describe()
```

```python
  # get high extreme
q1=25
q2=50
q3=75
whis=1.5
iq = q3 - q1
hi_val = q3 + whis * iq
wisk_hi = np.compress(df.Tweets <= hi_val, df.Tweets)
if len(wisk_hi) == 0 or np.max(wisk_hi) < q3:
    wisk_hi = q3
else:
    wisk_hi = max(wisk_hi)
# get low extreme
lo_val = q1 - whis * iq
wisk_lo = np.compress(df.Tweets >= lo_val, df.Tweets)
if len(wisk_lo) == 0 or np.min(wisk_lo) > q1:
    wisk_lo = q1
else:
    wisk_lo = min(wisk_lo)

print('q1=',wisk_lo)    
print ('q3=',wisk_hi)

```

```python
plt.boxplot(df.Tweets, notch=False, sym='+', vert=True, whis=1.5,
        positions=None, widths=None, patch_artist=False,
        bootstrap=None, usermedians=None, conf_intervals=None)
```

```python
quartiles=[]
df.loc[(df['Tweets'] >= 76) & (df['Tweets'] <= 100), 'quartiles'] = 4
df.loc[(df['Tweets'] >= 51) & (df['Tweets'] <= 75), 'quartiles'] = 3
df.loc[(df['Tweets'] >= 26) & (df['Tweets'] <= 50), 'quartiles'] = 2
df.loc[(df['Tweets'] >= 0) & (df['Tweets'] <= 25), 'quartiles'] = 1
   

df[quartiles]=quartiles
print(df)
```

```python
import seaborn as sns
sns.set(style="whitegrid")
#tips = sns.load_dataset("df")
ax = sns.boxplot(x="quartiles", y="Tweets", data=df)
```
